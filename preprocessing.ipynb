```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Crop_recommendation.csv")

# Basic info
print("Shape:", df.shape)
print("\nFirst 5 rows:\n", df.head())
print("\nMissing values:\n", df.isnull().sum())
print("\nCrop classes:", df['label'].nunique())
print(df['label'].value_counts())
print("\nStatistics:\n", df.describe())

# --- Plot 1: Crop distribution ---
plt.figure(figsize=(12, 5))
df['label'].value_counts().plot(kind='bar', color='steelblue')
plt.title("Number of Samples per Crop")
plt.xlabel("Crop")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("crop_distribution.png")
plt.show()

# --- Plot 2: Feature distributions ---
features = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
df[features].hist(figsize=(14, 8), bins=20, color='mediumseagreen', edgecolor='black')
plt.suptitle("Feature Distributions")
plt.tight_layout()
plt.savefig("feature_distributions.png")
plt.show()

# --- Plot 3: Correlation heatmap ---
plt.figure(figsize=(9, 6))
sns.heatmap(df[features].corr(), annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig("correlation_heatmap.png")
plt.show()

# --- Plot 4: Boxplot — NPK by crop ---
plt.figure(figsize=(16, 5))
for i, col in enumerate(['N', 'P', 'K']):
    plt.subplot(1, 3, i+1)
    sns.boxplot(data=df, x='label', y=col, palette='Set2')
    plt.xticks(rotation=90, fontsize=6)
    plt.title(f"{col} by Crop")
plt.tight_layout()
plt.savefig("npk_boxplot.png")
plt.show()

print("\nEDA complete! Charts saved.")
```
